creating daily dataset

```python
import pandas as pd

# Define file paths
input_path = r"C:\Users\msnin\Downloads\Oden\Oden-github\Measles Model\Datasets Used\MASTER DSHS DATA.csv"
output_path = r"C:\Users\msnin\Downloads\Oden\Oden-github\Measles Model\Datasets Used\daily_infections.csv"

# Read the CSV (fixes possible extra whitespaces, newlines)
df = pd.read_csv(input_path)

# Clean column names of whitespace and carriage return
df.columns = [c.strip() for c in df.columns]

# Convert date column, rename
df['Date'] = pd.to_datetime(df['Day of Epi Curve Date'], dayfirst=True, errors='coerce')

# Sort by date
start, end = df['Date'].min(), df['Date'].max()
df = df.sort_values('Date')

# Create all dates in interval
full_dates = pd.DataFrame({'Date': pd.date_range(start, end)})

# Merge with original data
daily = full_dates.merge(df[['Date', 'Confirmed']], on='Date', how='left')

# Fill NA confirmed values with 0
daily['Confirmed'] = daily['Confirmed'].fillna(0).astype(int)

# Keep required columns and save
daily[['Date', 'Confirmed']].to_csv(output_path, index=False)
```

weekly dataset

```python
import pandas as pd

# Define file paths
input_path = r"C:\Users\msnin\Downloads\Oden\Oden-github\Measles Model\Datasets Used\daily_infections.csv"
output_path = r"C:\Users\msnin\Downloads\Oden\Oden-github\Measles Model\Datasets Used\weekly_infections.csv"

# Read the daily infections CSV
df = pd.read_csv(input_path)

# Convert Date column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index to use resample
df.set_index('Date', inplace=True)

# Resample by week ending Saturday, summing the 'Confirmed' cases
weekly_df = df.resample('W-SAT').sum().reset_index()

# Save the weekly aggregated data
weekly_df.to_csv(output_path, index=False)
```

7 day rolling infections dataset

```python
import pandas as pd

# File paths
input_path = r"C:\Users\msnin\Downloads\Oden\Oden-github\Measles Model\Datasets Used\daily_infections.csv"
output_path = r"C:\Users\msnin\Downloads\Oden\Oden-github\Measles Model\Datasets Used\7_day_rolling_infections.csv"

# Read data
df = pd.read_csv(input_path)

# Convert Date to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Sort by Date (important for rolling window correctness)
df = df.sort_values('Date')

# Calculate 7-day centered rolling average on Confirmed cases
df['7_day_rolling_avg'] = df['Confirmed'].rolling(window=7, center=True, min_periods=1).mean()

# Save the result
df.to_csv(output_path, index=False)
```
